# GenAI V/S AgenticAI

- GenAI is about content creation, while agenticAI is about solving a goal.
- GenAI is reactive, while agenticAI is proactive(autonomous).
- GenAI is building block of agenticAI.
- AgenticAI is adaptive, generic, have memory, can take actions.
- GenAI is a capability where as agenticAI is a behaviour.


# AgenticAI

- AgenticAI is a type of AI that can take up a task or goal from user and then work toward completing it in its own, with minimal human guidance.
- It plans, takes action, adapts to changes and seeks help only when necessary.


## Key Characteristics of AgenticAI

- Autonomous
- Goal oriented
- Planning
- Reasoning
- Adaptability
- Context awareness

## AgenticAI Components

- Brain
- Memory
- Tools
- Orchestrator
- Supervisor

# Langchain V/S LangGraph

## Langchain Challenges:

- Control Flow Complexity
- Handling State
- Event Driven Execution
- Fault Tolerance
- Human In The Loop(HITL)
- Built for short running workflows
- Nested Workflows
- Observability : LangSmith


# LangGraph

- Langgraph is orchestration framework that enables you to build stateful, multistep and event driven workflows using LLM. its ideal for designing both single agent and multi agent ai applications.
- Think of langgraph as a flowchart engine for LLMs.
- Langgraph handles workflow orchestration , while langchain provides the building blocks for each step in that workflows.
