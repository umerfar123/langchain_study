# LangGraph

- Langgraph is orchestration framework that enables you to build stateful, multistep and event driven workflows using LLM. its ideal for designing both single agent and multi agent ai applications.
- Think of langgraph as a flowchart engine for LLMs.
- Langgraph handles workflow orchestration , while langchain provides the building blocks for each step in that workflows.
- It enables advanced features like parallelism, loops, branching, memory and resumability, making it ideal for agentic & production grade AI applications.

# Components / Core Concepts

- Nodes :
    - Every node in graph is task that is defined using python function.
- Edges :
    - Edges connects nodes which tell how two nodes are connected together and how to execute them.
    - Sequential Edges, Parallel Edges, Conditional Edges, Loops

- Each node represent the task to do and edges represents flow of execution

- State :
    - It is shared memory that flows through your workflow
    - Special type of dict that denotes current state of the graph.
    - Each node have access to state and can modifying them.
    - State is mutable.
    - Normally created using typedDict.

- Reducers :
    - It defines how updates from nodes are applied to the shared state.
    - Each key in state can have its own reducer, which determines whether new data replaces, merges, or adds to the existing value.
