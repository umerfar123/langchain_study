# Rise of LLMs

- Before LLMs era, the chatbot applications was not that good beacause the brain behind the application was not able to understand natural language properly **NLU**.
- Also the brain was failing in context aware text generation, **CATG**.
- Eventually when LLMs arrived it solved both this problem due to which chatbots become so popular.

# Langchain

- Founded by **Harrison Chase in 2022**.
- Langchain is an **open source framework** that helps in building LLM based applications.
- It provides **modular components and end-to-end tools** that help developers build complex AI applications such as chatbots, question answering systems, rag, autonomous agents and more.
- **Supports all the major LLMs**.
- Simplifies developing LLM based applications.
- Integration available for all major tools and services.
- Free and actively developed.
- Supports all major GenAI use cases.

### 1 - Why Langchain

- Concept of Chains : Creating complex pipelines.
- Model Agnostic development : Easy to use different llm models.
- Complete ecosytem : document loaders, embedding, prompts, 
- Memory and state building
