- Models
- Prompts

- Agents

## Models

- Models are the core interface through which you interact with AI models.
- LLMs generally are trained on whole internet data due to which they are heavy, we need a good hardware for running such LLMs.
- So the LLM developers started hosting the models on their server and created a API on top of this which anyone can hit and use this API for interacting with LLMs.
- But each LLM Providers had their on way using this API and communicating with LLMs which made difficulty for end users to track different such syntaxes.
- Thats where the model component of Langchain helps us by providing a common interface with which we can communicate different LLMs by changing very little much in the syntax.


### Types Foundational Models

- Langauge Models : i/p = text, o/p = text
- Embedding Models : i/p = text, o/p = vectors

## Prompts

- Prompts are the way in which we communicate with LLMs.
- Langchain provides Prompt templates which help in creating in dynamic and static prompts with which we can interact with LLMs efficiently.
- This template is suitable with LLMs providers.

## Chains

- Langchain name was formed from this component.
- Chain component allows us to create complex workflows or piplines easily by connecting different components like models, prompts, agents etc..
- Chain automatically passes o/p from one component to i/p of another component.
- E.g : Sequential Chain, Parallel Chain, Branch Chain etc..

## Indexes

- Indexes connect your application to external knowledge- such as PDFs, websites or databases.
- Document Loaders.
- Text Splitters.
- Vector Store.
- Retrievers.

## Memory

- LLM Api calls are stateless.
- LLM by default have no idea about your past conversation, past api calls.
- ConversationBufferMemory
- ConversationBufferWindowMemory
- Summarizer-Based Memory
- Custom Memory
